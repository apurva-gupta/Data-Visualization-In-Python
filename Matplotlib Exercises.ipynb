

# Matplotlib Exercises 


```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline.**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
axes = fig.add_axes([0.0,0.0,1.0,1.0])
axes.plot(x,y,color='blue',alpha=0.7)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0,0.0,1.0,1.0])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0,0.0,1.0,1.0])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax2.set_xlim([0,100])
ax2.set_ylim([0,200])
ax1.plot(x,y,color='red')
ax2.plot(x,y,color='red')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0,0.0,1.0,1.0])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0,0.0,1.0,1.0])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax2.set_xlim([20.0,22.0])
ax2.set_ylim([30,50])
ax1.plot(x,z,color='blue',alpha=0.6)
ax2.plot(x,y,color='blue',alpha=0.6)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color="blue",linestyle='--',linewidth=3)
axes[1].plot(x,z,color="red",linewidth=3)
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig,axes = plt.subplots(figsize=(12.5,2),nrows=1,ncols=2)
axes[0].plot(x,y,color="blue",linewidth=5)
axes[1].plot(x,z,color="red",ls='--',linewidth=2)
```
